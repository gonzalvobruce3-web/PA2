```python
import numpy as np
```

```python
np.random.seed(0)                 # reproducible results (optional)
X = np.random.random((5, 5))      # 5x5 ndarray with values in [0, 1)

mu = X.mean()                     # overall mean of X
sigma = X.std()                   # overall std of X
Z = (X - mu) / sigma              # normalized array
```

```python
print("===Normalization Problem===")
print("X:\n", X)
print("\nZ (normalized):\n", Z)
print("Z mean ≈ 0:", round(Z.mean(), 6), "| Z std ≈ 1:", round(Z.std(), 6))
print()
```

```python
A = (np.arange(1, 101) ** 2).reshape(10, 10)

div_by_3 = A[A % 3 == 0]          # boolean mask then 1D view of matching elements

print("=== Divisible by 3 Problem ===")
print("A (10x10 squares of 1..100):\n", A)
print("\nElements divisible by 3:")
print(div_by_3)
print("Count:", div_by_3.size)
```
